```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.arange(11)
y=x**2
print(x,y)
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(y,x,'y--')
```

```python
plt.scatter(x,y,)
```

```python
plt.plot(x,y)
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Figure-1/Plot')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r*--')

plt.subplot(1,3,2)
plt.plot(x,y,'g*--')

plt.subplot(1,3,3)
plt.plot(y,x,'y*--')

plt.tight_layout()
```

```python
fig1=plt.figure(figsize=(12,5))
axes=fig1.add_axes([.1,.1,.9,.9])
axes.plot(x,y,'r*--')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')


axes.set_xlim([1,4])
axes.set_ylim([0,50])

fig1.savefig("graph.png",dpi=200)
```

```python
fig,axes=plt.subplots(2,2)
plt.tight_layout()
```

```python
fig1,axes1=plt.subplots(2,1)


fig2,axes2=plt.subplots(1,2)
```

```python
axes1[0].plot(x,y)
fig1
```

```python
axes2[1].plot(x,y)
fig2
```

```python
axes2[0].plot(x,y,'b*')
axes2[0].set_xlabel('X_label')
axes2[0].set_ylabel('y_label')
axes2[0].set_title('Title-Blue ')

fig1.tight_layout()
axes2[1].plot(x,y,'r--')
axes2[1].set_xlabel('X_label')
axes2[1].set_ylabel('y_label')
axes2[1].set_title('Title-Red ')

fig2.tight_layout()
fig2
```

```python
fig3=plt.figure(figsize=(10,5))

axes_main=fig3.add_axes([0.1,0.1,0.9,0.9])

axes_inset=fig3.add_axes([.3,.5,.3,.4])


axes_main.plot(x,y,'b--')
axes_main.set_xlabel('X label')
axes_main.set_ylabel('Y label')
axes_main.set_title('Title')


axes_inset.plot(y,x,'r--')
axes_inset.set_xlabel('X label')
axes_inset.set_ylabel('Y label')
axes_inset.set_title('Title')
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

axes.plot(x,x**2 ,label='x**2')
axes.plot(x,x**3 ,label='x**3')
axes.legend()
```

```python
fig,axes=plt.subplots()
axes.plot(x,x+1,'g--')
axes.plot(x,x+2,'y--')
```

```python
fig, axes=plt.subplots()
axes.plot(x,x+2,'r--')
axes.plot(x,x+3,'b--')
axes.plot(x,x+4,'o--')
```
